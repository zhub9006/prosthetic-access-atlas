# Prosthetic Clinical Trial Trend Analysis
    

## Key Findings from ClinicalTrials.gov Data
- 155 prosthetic-related studies identified
- Only 21.3% are actively recruiting
- 34.2% completed — many without follow-up/reporting
- 33.5% have unknown phase — transparency gap
- <8% are Phase 3 or higher — late-stage evidence sparse
- 70.3% sponsored by academia — industry largely absent
- ZERO trials in rural Appalachia or Mississippi Delta
- U.S. dominates (48.5% of all global prosthetic trials)
- Developing nations nearly absent (India: 3, Pakistan: 2)


```python
import pandas as pd
import json

# Load the clinical trial data
df = pd.read_csv('data/clinical_trials.csv')

# Status distribution
with open('data/trial_trends_by_status.json') as f:
    status_data = json.load(f)

# Phase distribution
with open('data/trial_trends_by_phase.json') as f:
    phase_data = json.load(f)

# Regional distribution
with open('data/trial_trends_by_region.json') as f:
    region_data = json.load(f)

# Sponsor type distribution
with open('data/trial_trends_by_sponsor.json') as f:
    sponsor_data = json.load(f)

print('Status breakdown:', {k: v['count'] for k, v in status_data['results'].items()})
print('Phase breakdown:', {k: v['count'] for k, v in phase_data['results'].items()})
print('Top regions:', {k: v['count'] for k, v in list(region_data['results'].items())[:5]})
print('Sponsor breakdown:', {k: v['count'] for k, v in sponsor_data['results'].items()})
```
